```python
test_data = pd.read_csv('salary-test-mini.csv')
train_data = pd.read_csv('salary-train.csv')
```

```python
train_data.LocationNormalized.fillna('nan', inplace=True)
train_data.ContractTime.fillna('nan', inplace=True)
test_data.ContractTime.fillna('nan', inplace=True)
test_data.LocationNormalized.fillna('nan',inplace=True)
```

```python
dic = DictVectorizer()
```

```python
x_train_categ = dic.fit_transform(train_data[['LocationNormalized','ContractTime']].to_dict('records'))
x_test_categ = dic.transform(test_data[['LocationNormalized','ContractTime']].to_dict('records'))
```

```python
train_data.FullDescription =  train_data.FullDescription.map(lambda x: x.lower())
test_data.FullDescription = test_data.FullDescription.map(lambda x: x.lower())
```

```python
train_data.FullDescription = train_data.FullDescription.replace('[^a-zA-Z0-9]',' ',regex=True)
test_data.FullDescription = test_data.FullDescription.replace('[^a-zA-Z0-9]',' ',regex=True)
```

```python
tf_idf = TfidfVectorizer(min_df=5)
```

```python
x_train_text = tf_idf.fit_transform(train_data.FullDescription)
```

```python
x_test_text = tf_idf.transform(test_data.FullDescription)
```

```python
sample_matrix = sp.sparse.hstack([x_train_text,x_train_categ],)
#Не путать scypi.hstack и scypi.sparse.hstack 
```

```python
regressor = Ridge(alpha=1, random_state=241)
```

```python
regressor.fit(sample_matrix,train_data.SalaryNormalized)
```

```python
x_test_samples = sp.sparse.hstack([x_test_text,x_test_categ])
```

```python
test_pred = regressor.predict(x_test_samples)
```

```python
txt = open('subweek4_1.txt','w')
```

```python
txt.write('%0.2f %0.2f' % (test_pred[0],test_pred[1]))
txt.close()
```
